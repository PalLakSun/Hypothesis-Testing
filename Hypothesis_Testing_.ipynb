1. What is a null hypothesis (H₀) and why is it important in hypothesis testing?

Ans-A null hypothesis (denoted as H₀) is a fundamental concept in statistical hypothesis testing. It's a statement or assumption that there is no effect or no difference between groups, or that a relationship does not exist between the variables being studied. It essentially proposes that any observed effect in the data is due to chance or random variation, rather than a true underlying phenomenon.

Example:

If you're testing whether a new drug has an effect on blood pressure, the null hypothesis (H₀) might state that the drug has no effect on blood pressure ( any difference observed in blood pressure is due to random chance).

Why is the Null Hypothesis Important?

1.Starting Point for Testing: The null hypothesis provides a baseline or default assumption that you can either reject or fail to reject based on the data. It's a starting point for drawing conclusions about the population.

2.Objectivity: Hypothesis testing is about being objective and unbiased. The null hypothesis serves as a "neutral" position, against which any new evidence (data) can be tested. You test whether the evidence supports an alternative hypothesis (H₁), which is typically what you're trying to prove.

3.Decision Making: Hypothesis testing helps you make informed decisions. You use statistical tests to either reject the null hypothesis (if the evidence is strong enough) or fail to reject it (if the evidence isn't strong enough). This helps scientists and researchers decide whether their findings are meaningful or just random occurrences.

2. What does the significance level (α) represent in hypothesis testing?

Ans-The significance level (denoted as α, often referred to as alpha) is a threshold used in hypothesis testing to determine whether the observed data provides enough evidence to reject the null hypothesis. It essentially sets the standard for how much evidence you need before concluding that the observed effect is statistically significant.

What does α represent?

The significance level, α, represents the probability of making a Type I error, which means rejecting a true null hypothesis (a false positive). In other words, it's the risk you're willing to take that the results you're seeing are due to random chance rather than a real effect.

For example, if you set α = 0.05 (a common choice), you are allowing a 5% chance of incorrectly rejecting the null hypothesis when it's actually true. This is like saying, "I'm willing to accept that 5% of the time, I might get a false positive."

Interpreting the Alpha-

α = 0.05: This is the most commonly used significance level. It means that there is a 5% chance of rejecting the null hypothesis when it is actually true.

α = 0.01: This represents a more stringent test, where you are only willing to accept a 1% chance of making a Type I error.

α = 0.10: A less strict significance level, allowing for a 10% chance of Type I error. This might be used in exploratory studies where you're willing to take a higher risk of false positives.

-Rejecting the null hypothesis when it is actually true

-You think there is an effect or difference when there really isn't

-Probability: Equal to the significance level (α)

-Risk reduced by lowering α

-Failing to reject the null hypothesis when it is actually false

- You miss a real effect or difference

-Probability: Denoted by β (beta)

-Risk reduced by increasing sample size or power

4.Explain the difference between a one-tailed and two-tailed test. Give an example of each.

Ans-In hypothesis testing, one-tailed and two-tailed tests differ in how the alternative hypothesis is defined and where the rejection region lies in the sampling distribution.

A one-tailed test checks for an effect in only one specific direction (either greater than or less than).

Hypotheses

H₀ (null hypothesis): No effect or no difference

H₁ (alternative hypothesis): Effect exists in one direction only

When it's used

When you have a strong, justified reason to expect a result in a particular direction and results in the opposite direction are not of interest.

Example-

A teacher believes a new teaching method improves test scores.

H₀: The average score is ≤ 70

H₁: The average score is > 70

Here, the rejection region is only in the right tail of the distribution.

Two-Tailed Test

A two-tailed test checks for an effect in both directions—whether the value is significantly higher or lower than a specified value.

Hypotheses

H₀: No effect or no difference

H₁: There is an effect or difference (either direction)

When it's used

When you want to detect any change, regardless of direction, or when you do not have a strong directional expectation.

Example-

A manufacturer tests whether a machine’s average fill weight differs from the target of 500 g.

H₀: The average fill weight = 500 g

H₁: The average fill weight ≠ 500 g

Here, the rejection regions are in both tails of the distribution.

6.When should you use a Z-test instead of a t-test?

Ans-You should use a Z-test instead of a t-test when certain conditions about the population standard deviation and sample size are met. The key difference lies in how much information you have about the population.

Use a Z-test only when σ is known and the sample size is large. Otherwise, use a t-test.

Use a Z-test when:

1.The population standard deviation (σ) is known

-This is the most important condition.

-Z-tests assume σ is known, while t-tests use the sample standard deviation (s) as an estimate.

2.The sample size is large (usually n ≥ 30)

-With a large sample, the sampling distribution of the mean is approximately normal due to the Central Limit Theorem, even if the population itself is not normally distributed.

3.The population is normally distributed (or approximately so)

This condition is especially important for small samples.

For large samples, normality is less critical.

Use a t-test when:

1.The population standard deviation is unknown

-This is the most common real-world case.

2.The sample size is small (n < 30)

-The t-distribution accounts for extra uncertainty from estimating σ with s.
